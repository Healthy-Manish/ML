```python
# fetching series with help of conditions
mask= data['city']=='Hyderabad'
data[mask]
data[mask].shape #to count the numbers of total matches in hyderabad
```

```python
def get_city_match_count(city):
  mask = data['city']==city
  return data[mask].shape[0]


get_city_match_count('Rajkot')
```
